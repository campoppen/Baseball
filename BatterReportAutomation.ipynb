```python
def countPitches(pitch_type):
    pitches = len(df[df['TaggedPitchType'] == pitch_type])
    return pitches
```

```python
def calculateUsePercentage(pitch_type):
    pitches = len(df[df['TaggedPitchType'] == pitch_type])
    return round((pitches / total_pitches*100),1)
```

```python
def calculateAverageVelocity(pitch_type):
    avgVelo = df[df['TaggedPitchType'] == pitch_type]['RelSpeed'].mean()
    return round(avgVelo,2)
```

```python
def calculateStrikePercentage(pitch_type):
    pitches = len(df[df['TaggedPitchType'] == pitch_type])
    strikes = len(df[(df['TaggedPitchType'] == pitch_type) & (df['PitchCall'].isin(['StrikeSwinging', 'StrikeCalled', 'FoulBall', 'InPlay']))])
    return round(((strikes/pitches)*100 if pitches > 0 else 0),1)
```

```python
def calculateChasePercentage(pitch_type):
    if pitch_type is not None:
        pitch_df = df[df['TaggedPitchType'] == pitch_type]
    else:
        pitch_df = df

    # Strike zone dimensions in feet
    strike_zone_half_width = (17 + 2.9) / 24  # ~0.83 ft
    zone_top = 43 / 12  # 3.58 ft
    zone_bottom = 15 / 12  # 1.25 ft

    # Pitches outside the strike zone
    outside_zone = pitch_df[
        (pitch_df['PlateLocHeight'] < zone_bottom) |
        (pitch_df['PlateLocHeight'] > zone_top) |
        (pitch_df['PlateLocSide'] < -strike_zone_half_width) |
        (pitch_df['PlateLocSide'] > strike_zone_half_width)
    ]

    # Swings at pitches outside the zone (excluding bunts)
    swing_calls = ['StrikeSwinging', 'InPlay', 'FoulBallNotFieldable', 'FoulBallFieldable']
    chases = outside_zone[
        outside_zone['PitchCall'].isin(swing_calls) &
        (outside_zone['TaggedHitType'] != 'Bunt')
    ]

    total_outside = len(outside_zone)
    if total_outside == 0:
        return "NA"
    
    return round((len(chases) / total_outside) * 100, 1)
```

```python
def calculateWhiffPercentage(pitch_type):
    if pitch_type is not None:
        pitch_df = df[df['TaggedPitchType'] == pitch_type]
    else:
        pitch_df = df

    # Filter to swinging events, excluding bunts
    swings = pitch_df[
        (pitch_df['PitchCall'].isin([
            'StrikeSwinging', 'InPlay', 'FoulBallNotFieldable', 'FoulBallFieldable'
        ])) &
        (pitch_df['TaggedHitType'] != 'Bunt')
    ]

    total_swings = len(swings)
    if total_swings == 0:
        return "NA"

    whiffs = len(swings[swings['PitchCall'] == 'StrikeSwinging'])
    return round((whiffs / total_swings) * 100, 1)
```

```python
def calculateHardHit(pitch_type):
    # Filter for balls in play excluding bunts
    in_play_df = df[
        (df['TaggedPitchType'] == pitch_type) & 
        (df['PitchCall'] == 'InPlay') & 
        (df['TaggedHitType'] != 'Bunt')  # Exclude bunts
    ]
    total_hits = len(in_play_df[in_play_df['ExitSpeed'].notna()])
    hard_hits = len(in_play_df[in_play_df['ExitSpeed'] > 90])
    
    if total_hits == 0:
        return "NA"  # Return "NA" if no balls were in play
    return round((hard_hits / total_hits) * 100, 1)
```

```python
def calculateExitVelo(pitch_type):
    in_play_df = df[
        (df['TaggedPitchType'] == pitch_type) & 
        (df['PitchCall'] == 'InPlay') & 
        (df['TaggedHitType'] != 'Bunt')  # Exclude bunts
    ]
    if in_play_df.empty:
        return "NA"  # Return "NA" if no balls were in play
    velo = in_play_df['ExitSpeed'].mean()
    return "NA" if pd.isna(velo) else round(velo, 2)
```

```python
def calculateTwoStrikeUsage(pitch_type):
    filtered_df = df[
        (df['Strikes'] == 2) &
        (df['Balls'].isin([0, 1, 2]))
    ]

    total_pitches = len(filtered_df)

    if pitch_type is None or total_pitches == 0:
        return "NA"

    pitch_count = (filtered_df['TaggedPitchType'] == pitch_type).sum()
    return round((pitch_count / total_pitches) * 100, 1)
```

```python
def calculateTwoStrikeChasePitch(pitch_type):
    # Filter to 2-strike count with 0-2 balls
    filtered_df = df[
        (df['Strikes'] == 2) &
        (df['Balls'].isin([0, 1, 2]))
    ]

    if pitch_type is not None:
        filtered_df = filtered_df[filtered_df['TaggedPitchType'] == pitch_type]

    # Strike zone dimensions in feet
    strike_zone_half_width = (17 + 2.9) / 24  # ~0.83 ft
    zone_top = 43 / 12  # 3.58 ft
    zone_bottom = 15 / 12  # 1.25 ft

    # Pitches outside the strike zone
    outside_zone = filtered_df[
        (filtered_df['PlateLocHeight'] < zone_bottom) |
        (filtered_df['PlateLocHeight'] > zone_top) |
        (filtered_df['PlateLocSide'] < -strike_zone_half_width) |
        (filtered_df['PlateLocSide'] > strike_zone_half_width)
    ]

    # Swinging events (excluding bunts)
    swing_calls = ['StrikeSwinging', 'InPlay', 'FoulBallNotFieldable', 'FoulBallFieldable']
    chases = outside_zone[
        outside_zone['PitchCall'].isin(swing_calls) &
        (outside_zone['TaggedHitType'] != 'Bunt')
    ]

    total_outside = len(outside_zone)
    if total_outside == 0:
        return "NA"
    
    return round((len(chases) / total_outside) * 100, 1)
```

```python
def calculateTwoStrikeWhiffPitch(pitch_type):
    # Filter for 2-strike counts with 0–2 balls
    filtered_df = df[
        (df['Strikes'] == 2) &
        (df['Balls'].isin([0, 1, 2]))
    ]

    # Apply pitch type filter if provided
    if pitch_type is not None:
        filtered_df = filtered_df[filtered_df['TaggedPitchType'] == pitch_type]

    # Filter to swinging events, excluding bunts
    swings = filtered_df[
        (filtered_df['PitchCall'].isin([
            'StrikeSwinging', 'InPlay', 'FoulBallNotFieldable', 'FoulBallFieldable'
        ])) &
        (filtered_df['TaggedHitType'] != 'Bunt')
    ]

    total_swings = len(swings)
    if total_swings == 0:
        return "NA"

    whiffs = (swings['PitchCall'] == 'StrikeSwinging').sum()
    return round((whiffs / total_swings) * 100, 1)
```

```python
def calculatePlateAppearances():
    return df[["GameID", "Inning", "PAofInning"]].drop_duplicates().shape[0]
```

```python
def calculateAtBats():
    # Step 1: Identify unique plate appearances
    pa_df = df[["GameID", "Inning", "PAofInning", "KorBB", "PitchCall", "PlayResult", "TaggedHitType"]].copy()

    # Step 2: Keep only the last pitch of each PA
    pa_df["PitchNo"] = df["PitchNo"]
    last_pitches = pa_df.sort_values("PitchNo").drop_duplicates(subset=["GameID", "Inning", "PAofInning"], keep="last")

    # Step 3: Identify excluded events
    is_walk = last_pitches["KorBB"].isin(["Walk"])
    is_hbp = last_pitches["PitchCall"].isin(["HitByPitch"])
    is_sac_bunt = last_pitches["PlayResult"].str.contains("Sacrifice", na=False) & (last_pitches["TaggedHitType"] == "Bunt")
    is_sac_fly = last_pitches["PlayResult"].str.contains("Sacrifice", na=False) & (last_pitches["TaggedHitType"] == "FlyBall")
    excluded = is_walk | is_hbp | is_sac_bunt | is_sac_fly

    # Step 4: Keep only valid ABs
    valid_abs = last_pitches[~excluded]

    return valid_abs.shape[0]
```

```python
def calculateWalks():
    return len(df[df['KorBB'] == 'Walk'])
```

```python
def calculateHits():
    return len(df[df['PlayResult'].isin(['Single', 'Double', 'Triple', 'HomeRun'])])
```

```python
def calculateStrikeOuts():
    return len(df[df['KorBB'] == 'Strikeout'])
```

```python
def calculateHBP():
    return len(df[df['PitchCall'] == 'HitByPitch'])
```

```python
def calculateRBIs():
    prev_runs = df["RunsScored"].shift(1).fillna(0)
    runs_scored = df["RunsScored"] > prev_runs
    is_hit = df["TaggedHitType"].isin(["GroundBall", "LineDrive", "FlyBall"])
    is_sac_fly = (df["TaggedHitType"] == "FlyBall") & (df["PlayResult"].str.contains("Out", na=False))
    is_sac_bunt = (df["TaggedHitType"] == "Bunt") & (df["PlayResult"].str.contains("Out", na=False)) & runs_scored
    is_bb_rbi = (df["KorBB"] == "Walk") & runs_scored
    is_hbp_rbi = (df["PitchCall"] == "HitByPitch") & runs_scored
    is_error = df["PlayResult"].str.contains("Error", na=False)
    is_double_play = df["PlayResult"].str.contains("Double Play", na=False)
    rbi_plays = runs_scored & (is_hit | is_sac_fly | is_sac_bunt | is_bb_rbi | is_hbp_rbi) & ~is_error & ~is_double_play
    return rbi_plays.sum()
```

```python
def calculateSingles():
    is_single = df["PlayResult"].str.contains("Single", na=False)
    return int(is_single.sum())
```

```python
def calculateDoubles():
    is_double = df["PlayResult"].str.contains("Double", na=False)
    return int(is_double.sum())
```

```python
def calculateTriples():
    is_triple = df["PlayResult"].str.contains("Triple", na=False)
    return int(is_triple.sum())
```

```python
def calculateHomeRuns():
    is_hr = df["PlayResult"].str.contains("HomeRun", na=False)
    return int(is_hr.sum())
```

```python
def calculateAverage():
    return calculateHits()/calculateAtBats()
```

```python
def calculateOBP():
    hits = calculateHits()
    walks = calculateWalks()
    hbp = len(df[df["PitchCall"] == "HitByPitch"])  # Hit By Pitch
    ab = calculateAtBats()
    sf = len(df[df["PlayResult"].str.contains("Sacrifice", na=False) & (df["TaggedHitType"] == "FlyBall")])  # Sac Flies
    denominator = ab + walks + hbp + sf
    if denominator == 0:
        return None  # Avoid division by zero
    return ((hits + walks + hbp) / denominator)
```

```python
def calculateSlugging():
    weightedDoubles = 2*calculateDoubles()
    weightedTriples = 3*calculateTriples()
    weightedHRs = 4*calculateHomeRuns()
    return (calculateSingles()+weightedDoubles+weightedTriples+weightedHRs)/calculateAtBats()
```

```python
def calculateOPS():
    return calculateSlugging() + calculateOBP()
```

```python
game_stats = {
        'PA': calculatePlateAppearances(),
        'AB': calculateAtBats(),
        'H': calculateHits(),
        'RBI': calculateRBIs(),
        'BB': calculateWalks(),
        'K': calculateStrikeOuts(),
        '1B': calculateSingles(),
        '2B': calculateDoubles(),
        '3B': calculateTriples(),
        'HR': calculateHomeRuns(),
        'AVG': calculateAverage(),
        'OBP': calculateOBP(),
        'SLG': calculateSlugging(),
        'OPS': calculateOPS()
}
```

```python
def draw_strike_zone(ax):
    strike_zone_width = (17+2.9)/12  # plate + ball
    strike_zone_height = [15/12, 43/12] # 6'
    buffer = 2.9/12

    # Dashed standard strike zone (17 inches wide)
    standard_zone = plt.Rectangle(
        (-strike_zone_width / 2, strike_zone_height[0]),
        strike_zone_width,
        strike_zone_height[1] - strike_zone_height[0],
        fill=False,
        color='black',
        linestyle='-',
        linewidth=2
    )
    ax.add_patch(standard_zone)

    # Solid extended strike zone (with ball buffer)
    extended_zone = plt.Rectangle(
        (-strike_zone_width / 2 - buffer, strike_zone_height[0] - buffer),
        strike_zone_width + 2 * buffer,
        (strike_zone_height[1] - strike_zone_height[0]) + 2 * buffer,
        fill=False,
        color='black',
        linestyle='--',
        linewidth=2
    )
    ax.add_patch(extended_zone)

    # Home plate line
    ax.plot([-(strike_zone_width-buffer) / 2, (strike_zone_width-buffer) / 2], [0, 0], 'k-', linewidth=2)
```

```python
def create_spray_chart(df):
    spray_df = df.copy()
    spray_df = spray_df[['Bearing', 'Distance', 'PlayResult']].dropna()
    valid_results = ['Out', 'Single', 'Double', 'Triple', 'HomeRun', 'Sacrifice', 'FieldersChoice','Error']
    spray_df = spray_df[spray_df['PlayResult'].isin(valid_results)]

    # Return empty plot if no data
    fig, ax = plt.subplots(figsize=(6, 6))

    # Convert to Cartesian coordinates
    spray_df['AngleRad'] = np.radians(spray_df['Bearing'])
    spray_df['x'] = spray_df['Distance'] * np.sin(spray_df['AngleRad'])
    spray_df['y'] = spray_df['Distance'] * np.cos(spray_df['AngleRad'])

    # Color scheme
    result_colors = {
        'Out': 'black',
        'Single': 'red',
        'Double': 'orange',
        'Triple': 'yellow',
        'HomeRun': 'green',
        'Sacrifice': 'blue',
        'FieldersChoice': 'purple',
        'Error': '#D3D3D3'
    }

    angles_deg = np.linspace(-45, 45, 500)
    angles_rad = np.radians(angles_deg)
    distances = 325 + 10 * np.cos(angles_rad)
    x_arc = distances * np.sin(angles_rad)
    y_arc = distances * np.cos(angles_rad)

    # Infield arc (130 at corners to 155 at center)
    infield_distances = 130 + 25 * np.cos(angles_rad)
    x_infield = infield_distances * np.sin(angles_rad)
    y_infield = infield_distances * np.cos(angles_rad)

    # Fill infield (dirt area)
    infield_polygon = np.column_stack(([0, *x_infield, 0], [0, *y_infield, 0]))
    ax.fill(infield_polygon[:, 0], infield_polygon[:, 1], color='saddlebrown', alpha=0.3, zorder=0)

    # Fill outfield (grass area)
    outfield_polygon = np.column_stack(([0, *x_arc[::-1], *x_infield, 0], [0, *y_arc[::-1], *y_infield, 0]))
    ax.fill(outfield_polygon[:, 0], outfield_polygon[:, 1], color='green', alpha=0.3, zorder=0)

    # Plot batted balls
    for result in valid_results:
        subset = spray_df[spray_df['PlayResult'] == result]
        ax.scatter(subset['x'], subset['y'],
                   color=result_colors.get(result, 'black'),
                   label=result, alpha=0.7, edgecolor='k', s=40)

    # Draw outfield arc and foul lines
    ax.plot(x_arc, y_arc, color='black')
    ax.plot([0, x_arc[0]], [0, y_arc[0]], color='black')  # Left foul line
    ax.plot([0, x_arc[-1]], [0, y_arc[-1]], color='black')  # Right foul line

    # Formatting
    ax.set_xlim(-250, 250)
    ax.set_ylim(0, 500)
    ax.set_aspect('equal')
    ax.set_title("Spray Chart", fontsize=14)
    ax.legend(
        title="Play Result",
        bbox_to_anchor=(1.05, 1),
        loc='upper left',
        fontsize=11,
        title_fontsize=12)
    ax.axis('off')

    plt.tight_layout()
    plt.close(fig)
    return fig
```

```python
def create_pitch_location(df):
    fig, ax = plt.subplots(figsize=(6, 6))
    
    if not df.empty:
        pitch_counts = df["TaggedPitchType"].value_counts()
        sorted_pitches = pitch_counts.index.tolist()

        for pitch_type in sorted_pitches:
            subset = df[df["TaggedPitchType"] == pitch_type]
            ax.scatter(subset["PlateLocSide"], subset["PlateLocHeight"], 
                       label=pitch_type, alpha=0.8, s=144,
                       color=pitch_colors.get(pitch_type, "gray"))

        draw_strike_zone(ax)  # Add strike zone

        ax.set_title("Pitch Location (Batter's Perspective)", fontsize=18)
        ax.set_xlabel("Horizontal Location (ft)", fontsize=14)
        ax.set_ylabel("Height (ft)", fontsize=14)
        ax.set_xlim(3, -3)
        ax.set_ylim(-1, 6)
    plt.tight_layout()
    plt.close(fig)
    return fig
```

```python
def create_all_pitches_heatmap(df):
    batter_data = df.copy()
    batter_data = batter_data.dropna(subset=['PlateLocSide', 'PlateLocHeight'])

    if batter_data.empty or len(batter_data) < 3:
        return None

    fig, ax = plt.subplots(figsize=(6, 6))

    sns.kdeplot(
        data=batter_data,
        x='PlateLocSide',
        y='PlateLocHeight',
        fill=True,
        cmap='RdBu_r',
        ax=ax
    )

    draw_strike_zone(ax)
    ax.set_title('All Pitches', fontsize=18)
    ax.set_xlabel('Horizontal Location (ft)', fontsize=15)
    ax.set_ylabel('Height (ft)', fontsize=15)
    ax.set_xlim(3, -3)  # Flip x-axis
    ax.set_ylim(-1, 6)
    ax.tick_params(axis='both', which='major', labelsize=10)

    plt.tight_layout()
    plt.close(fig)
    return fig
```

```python
def create_density_plot(df):
    # Drop missing values
    batter_data = df.dropna(subset=['PlateLocSide', 'PlateLocHeight', 'ExitSpeed'])

    # Filter for Exit Speed ≥ 90 mph
    batter_data = batter_data[batter_data['ExitSpeed'] >= 90]

    if batter_data.empty or len(batter_data) < 3:
        return None

    # Create the heatmap
    fig, ax = plt.subplots(figsize=(6, 6))

    sns.kdeplot(
        data=batter_data,
        x='PlateLocSide',
        y='PlateLocHeight',
        weights=batter_data['ExitSpeed'],
        fill=True,
        cmap='RdBu_r',
        ax=ax
    )

    draw_strike_zone(ax)

    # Formatting
    ax.set_title('90+ Exit Velo', fontsize=18)
    ax.set_xlabel('Horizontal Location (ft)', fontsize=14)
    ax.set_ylabel("Height (ft)", fontsize=14)
    ax.set_xlim(3, -3)
    ax.set_ylim(-1, 6)
    ax.tick_params(axis='both', which='major', labelsize=10)

    plt.tight_layout()
    plt.close(fig)
    return fig
```

```python
def highlight_metrics(val, metric):
        
        # Average
        if metric == 'AVG':
            if val >= 0.330:
                return '#008000'  # Dark Green (Really Good)
            elif val >= 0.300:
                return '#90EE90'  # Light Green (Good)
            elif val < 0.240:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val < 0.270:
                return '#FF7F7F'  # Light Red (Bad)
        
        # On-base percentage
        elif metric == 'OBP':
            if val >= 0.400:
                return '#008000'  # Dark Green (Really Good)
            elif val >= 0.370:
                return '#90EE90'  # Light Green (Good)
            elif val < 0.310:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val < 0.340:
                return '#FF7F7F'  # Light Red (Bad)
        
        # Slugging percentage
        elif metric == 'SLG':
            if val >= 0.500:
                return '#008000'  # Dark Green (Really Good)
            elif val >= 0.450:
                return '#90EE90'  # Light Green (Good)
            elif val < 0.400:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val < 0.350:
                return '#FF7F7F'  # Light Red (Bad)
        
        # On-base plus slugging
        elif metric == 'OPS':
            if val >= 0.900:
                return '#008000'  # Dark Green (Really Good)
            elif val >= 0.850:
                return '#90EE90'  # Light Green (Good)
            elif val < 0.700:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val < 0.750:
                return '#FF7F7F'  # Light Red (Bad)
        
        # Chase %
        elif metric in ['Chase%', '2kChase%']:
            if val >= 35:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val >= 30:
                return '#FF7F7F'  # Light Red (Bad)
            elif val < 20:
                return '#008000'  # Dark Green (Really Good)
            elif val < 25:
                return '#90EE90'  # Light Green (Good)
        
        # Whiff %
        elif metric in ['Whiff%', '2kWhiff%']:
            if val >= 30:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val >= 25:
                return '#FF7F7F'  # Light Red (Bad)
            elif val < 15:
                return '#008000'  # Dark Green (Really Good)
            elif val < 20:
                return '#90EE90'  # Light Green (Good)

        # Hard Hit %
        elif metric == 'HardHit%':
            if val >= 50:
                return '#008000'  # Dark Green (Really Good)
            elif val >= 40:
                return '#90EE90'  # Light Green (Good)
            elif val < 20:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val < 30:
                return '#FF7F7F'  # Light Red (Bad)

        # Average exit velo
        elif metric == 'AvgExitVelo':
            if val >= 95:
                return '#008000'  # Dark Green (Really Good)
            elif val >= 90:
                return '#90EE90'  # Light Green (Good)
            elif val < 80:
                return '#FF0000'  # Dark Red (Really Bad)
            elif val < 85:
                return '#FF7F7F'  # Light Red (Bad)
        
        # Otherwise, return nothing
        return ''
```

```python
def create_batter_report(batter_name, pitch_stats_df, game_stats_df, df):
    
    # Find most common BatterTeam
    top_team = game_df[game_df['PitcherTeam'] != 'POI_LOM']['PitcherTeam'].mode().iloc[0]
    new_team = top_team.replace(" ", "_")

    # Build the path: Reports/<OpponentTeam>/
    reports_folder = os.path.join("Reports")

    # Ensure the team-specific folder exists
    os.makedirs(reports_folder, exist_ok=True)

    # Clean and reorder pitcher name from "Lastname, Firstname" to "FirstnameLastname"
    last, first = batter_name.split(", ")
    formatted_batter_name = f"{first}{last}".replace(" ", "")
    title_batter_name = f"{first} {last}"

    # Create filename
    output_filename = os.path.join(reports_folder, f"{formatted_batter_name}_vs_{new_team}.pdf")

    doc = SimpleDocTemplate(
        output_filename,
        pagesize=landscape(letter),
        rightMargin=10,
        leftMargin=10,
        topMargin=10,
        bottomMargin=10
    )

    elements = []

    styles = getSampleStyleSheet()
    title_style = ParagraphStyle('CustomTitle', parent=styles['Heading1'], fontSize=24, spaceAfter=7.5, alignment=1, textColor=PLNU_GREEN)
    heading_style = ParagraphStyle('CustomHeading', parent=styles['Heading2'], textColor=PLNU_GREEN)

    elements.append(Paragraph("Point Loma Nazarene University", title_style))
    elements.append(Paragraph(f"Batter Analysis Report: {title_batter_name}", title_style))
    elements.append(Spacer(1, 1))

    # Game stats
    elements.append(Paragraph("Game Statistics", heading_style))
    elements.append(Spacer(1, 1))

    for col in ['AVG', 'OBP', 'SLG', 'OPS']:
        if col in game_stats_df.columns:
            game_stats_df[col] = game_stats_df[col].round(3)

    game_styles = []
    game_data = [list(game_stats_df.columns)]
    for row_idx, row in enumerate(game_stats_df.itertuples(index=False), start=1):
        game_row = []
        for col_idx, (col, val) in enumerate(zip(game_stats_df.columns, row)):
            if isinstance(val, float):
                highlight_color = highlight_metrics(val, col)
                game_row.append(f"{val:.3f}")
                if highlight_color:
                    game_styles.append(('BACKGROUND', (col_idx, row_idx), (col_idx, row_idx), highlight_color))
            else:
                game_row.append(str(val))
        game_data.append(game_row)

    game_col_widths = [doc.width / len(game_data[0])] * len(game_data[0])
    game_table = Table(game_data, colWidths=game_col_widths)
    game_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), PLNU_GREEN),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 10),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 5),
        ('BACKGROUND', (0, 1), (-1, -1), colors.white),
        ('TEXTCOLOR', (0, 1), (-1, -1), colors.black),
        ('FONTNAME', (0, 1), (-1, -1), 'Helvetica'),
        ('FONTSIZE', (0, 1), (-1, -1), 9),
        ('GRID', (0, 0), (-1, -1), 1, PLNU_GREEN),
        ('BOX', (0, 0), (-1, -1), 2, PLNU_GREEN),
        ('LINEBELOW', (0, 0), (-1, 0), 2, PLNU_GOLD),
    ] + game_styles))
    elements.append(game_table)

    # Pitch stats
    elements.append(Paragraph("Pitch Statistics", heading_style))
    elements.append(Spacer(1, 1))

    pitch_stats_df = pitch_stats_df.rename(columns={
        'total_pitches': 'Total',
        'use_percentage': 'Usage%',
        'strike_percentage': 'Strike%',
        'chase_percentage': 'Chase%',
        'whiff_percentage': 'Whiff%',
        'hard_hit_percentage': 'HardHit%',
        'avg_exit_velo': 'AvgExitVelo',
        'two_strike_usage': '2kUsage%',
        'two_strike_chase': '2kChase%',
        'two_strike_whiff': '2kWhiff%'
    })

    pitch_styles = []
    numeric_columns = ['AvgExitVelo', 'Usage%', 'Strike%', 'Chase%', 'Whiff%', 'HardHit%', '2kUsage%', '2kWhiff%', '2kChase%']
    pitch_stats_df[numeric_columns] = pitch_stats_df[numeric_columns].apply(
        lambda col: pd.to_numeric(col, errors='coerce').where(col != "NA", other="NA")
    )

    pitch_data = [['Pitch'] + list(pitch_stats_df.columns)]
    for row_idx, (pitch_name, row) in enumerate(pitch_stats_df.iterrows(), start=1):
        pitch_row = [pitch_name]
        for col_idx, (col, val) in enumerate(row.items(), start=1):
            if isinstance(val, float):
                highlight_color = highlight_metrics(val, col)
                val_str = f"{val:.1f}" if col in ['AvgExitVelo'] or col.endswith('%') else str(int(val))
                pitch_row.append(val_str)
                if highlight_color:
                    pitch_styles.append(('BACKGROUND', (col_idx, row_idx), (col_idx, row_idx), highlight_color))
            else:
                pitch_row.append(str(val))
        pitch_data.append(pitch_row)

    col_widths = [1.2 * inch] + [(doc.width - 1.2 * inch) / (len(pitch_data[0]) - 1)] * (len(pitch_data[0]) - 1)
    pitch_table = Table(pitch_data, colWidths=col_widths)
    pitch_table.setStyle(TableStyle([
        ('BACKGROUND', (0, 0), (-1, 0), PLNU_GREEN),
        ('TEXTCOLOR', (0, 0), (-1, 0), colors.white),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
        ('FONTSIZE', (0, 0), (-1, 0), 9),
        ('BOTTOMPADDING', (0, 0), (-1, 0), 5),
        ('BACKGROUND', (0, 1), (-1, -1), colors.white),
        ('TEXTCOLOR', (0, 1), (-1, -1), colors.black),
        ('FONTNAME', (0, 1), (-1, -1), 'Helvetica'),
        ('FONTSIZE', (0, 1), (-1, -1), 9),
        ('GRID', (0, 0), (-1, -1), 1, PLNU_GREEN),
        ('BOX', (0, 0), (-1, -1), 2, PLNU_GREEN),
        ('LINEBELOW', (0, 0), (-1, 0), 2, PLNU_GOLD),
    ] + pitch_styles))
    elements.append(pitch_table)

    # ---- Bottom Layout: Spray Chart | Legend | Pitch Location ----
    spray_chart_fig = create_spray_chart(df)
    pitch_location_fig = create_pitch_location(df)

    spray_chart_img = Image(save_plt_as_png(spray_chart_fig), width=4.4 * inch, height=3.25 * inch) if spray_chart_fig else Spacer(1, 3.25 * inch)
    pitch_location_img = Image(save_plt_as_png(pitch_location_fig), width=3.25 * inch, height=3.25 * inch) if pitch_location_fig else Spacer(1, 3.25 * inch)

    sorted_pitches = pitch_stats_df.sort_values(by="Total", ascending=False).index.tolist()
    legend_data = [["Pitch Types"]] + [[Paragraph(f'<font color="{pitch_colors.get(pitch, "gray")}">●</font>&nbsp;&nbsp;&nbsp;{pitch}', styles["Normal"])] for pitch in sorted_pitches]
    legend_table = Table(legend_data, colWidths=[1.25 * inch])
    legend_table.setStyle(TableStyle([
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER')
    ]))

    bottom_row = [spray_chart_img, pitch_location_img, legend_table]
    bottom_table = Table([bottom_row], colWidths=[4.75 * inch, 3.75 * inch, 1.5 * inch])
    bottom_table.setStyle(TableStyle([
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('TOPPADDING', (0, 0), (-1, -1), 10)
    ]))

    elements.append(Spacer(1, 1))
    elements.append(bottom_table)
    elements.append(PageBreak())

    # ---- Page 2 Placeholder ----
    elements.append(Paragraph("Heatmaps (Batter's Perspective)", heading_style))
    
    all_pitches_heatmap_fig = create_all_pitches_heatmap(df)
    fastball_heatmap_fig = create_fastballs_heatmap(df)
    offspeed_heatmap_fig = create_offspeed_heatmap(df)
    density_plot_fig = create_density_plot(df)
    two_strike_fig = create_two_strike_heatmap(df)
    two_strike_offspeed_fig = create_two_strike_offspeed_heatmap(df)

    all_pitches_heatmap_img = Image(save_plt_as_png(all_pitches_heatmap_fig), width=3.25 * inch, height=3.5 * inch) if all_pitches_heatmap_fig else Spacer(1, 3.5 * inch)
    fastball_heatmap_img = Image(save_plt_as_png(fastball_heatmap_fig), width=3.25 * inch, height=3.5 * inch) if fastball_heatmap_fig else Spacer(1, 3.5 * inch)
    offspeed_heatmap_img = Image(save_plt_as_png(offspeed_heatmap_fig), width=3.25 * inch, height=3.5 * inch) if offspeed_heatmap_fig else Spacer(1, 3.5 * inch)
    density_plot_img = Image(save_plt_as_png(density_plot_fig), width=3.25 * inch, height=3.5 * inch) if density_plot_fig else Spacer(1, 3.5 * inch)
    two_strike_img = Image(save_plt_as_png(two_strike_fig), width=3.25 * inch, height=3.5 * inch) if two_strike_fig else Spacer(1, 3.5 * inch)
    two_strike_offspeed_img = Image(save_plt_as_png(two_strike_offspeed_fig), width=3.25 * inch, height=3.5 * inch) if two_strike_offspeed_fig else Spacer(1, 3.5 * inch)
    
    # Arrange images in a 3-column x 2-row layout
    heatmap_grid_data = [
        [all_pitches_heatmap_img, fastball_heatmap_img, offspeed_heatmap_img],
        [density_plot_img, two_strike_img, two_strike_offspeed_img]
    ]

    heatmap_grid = Table(heatmap_grid_data, colWidths=[3.25 * inch] * 3)
    heatmap_grid.setStyle(TableStyle([
        ('VALIGN', (0, 0), (-1, -1), 'MIDDLE'),
        ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
        ('TOPPADDING', (0, 0), (-1, -1), 10),
        ('BOTTOMPADDING', (0, 0), (-1, -1), 10),
    ]))

    # Add the table to the PDF elements
    elements.append(heatmap_grid)

    doc.build(elements)
```
